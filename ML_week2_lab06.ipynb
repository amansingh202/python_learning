# Using Scikit-Learn library for Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from lab_utils_multi import load_house_data
plt.style.use('./deeplearning.mplstyle')
np.set_printoptions(precision = 2)
```

```python
#load the data set
X_train = np.array([1, 2])
y_train = np.array([300, 500])
```

```python
#create and fit the model
linear_model = LinearRegression()

#X must be a 2D matrix

linear_model.fit(X_train.reshape(-1, 1), y_train)
```

```python
b = linear_model.intercept_
w = linear_model.coef_
print(f"w : {w:}, b = {b:0.2f}")
print(f"Manual prediction is (wx +b) is :{w*1200 + b}")
```

# Multiple features (Regression)

```python
X_train, y_train = load_house_data()
X_features = ['size(sqft)', 'bedrooms', 'floors', 'age']
```

```python
#using Linear Regression method from sklear.leanr_model
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)
```

```python
#w and b parameters
b = linear_model.intercept_
w = linear_model.coef_
print(f"w = {w:}, b = {b:0.2f}")
```

```python
#predicting the output now
print(f"Prediction on trainig set:\n {linear_model.predict(X_train)[:4]}")
print(f"Prediction using manual w,b is wx+b: {X_train @ w +b}[:4]")

print(f"Traget values \n {y_train[:4]}")

x_house = np.array([1200, 3, 1, 40]).reshape(-1,4)
x_house_predict = linear_model.predict(x_house)[0]

print(f"predicted value of a house with 1200 sqft area, 3 bedrooms, 1 floor and 40 years old is: ${x_house_predict*1000 + b}")
```
