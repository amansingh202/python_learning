# reshaping layout

```python
import pandas as pd
```

```python
df = pd.read_csv("population_of_world.csv")
df
```

```python
df.sort_values(by = "2022 Population", ascending = False)
df
```

```python

#filter for population > 100,000,000 in 2022
df_2022pop = df[df["2022 Population"] > 100000000]
df_2022pop
```

```python
#reshape the table
#long to wide table format
df_subset = df_2022pop.sort_index().groupby(["2022 Population"])
df_subset.head(20)
```

```python
#values next to each other 
df.pivot(columns = "Continent", values = "2022 Population")
```

```python
df.pivot(columns = "Continent", values = "2022 Population").plot()
```

```python
df.pivot_table(values = "2022 Population", index = "Country", columns = "Continent", aggfunc = "mean",margins = True)
```
