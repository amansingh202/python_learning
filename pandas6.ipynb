```python
import pandas as pd
```

# create new columns from the existing columns

```python
df = pd.read_csv("population_of_world.csv")
df
```

```python
#creating a new column 1800 population assuming it to be one third of the 1970 population
df["1800 population(assume)"] = df["1970 Population"] * 0.3
df.head()
```

```python
#renaming a column
df_renamed = df.rename(columns= {"Growth Rate":"Rate of Growth",
                   "World Population Percentage": "World Population %"})
df_renamed
```

```python
#converting columns name in lower case
df_renamed_lowercase = df.rename(columns = str.lower)
df_renamed_lowercase
```
