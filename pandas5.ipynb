# creating plots in pandas

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
#this matplot lib is used for making plot
```

```python
popList = pd.read_excel("world_population_excel_workbook.xlsx")
```

```python
popList.head(6)
```

```python
top5 = popList.loc[popList["1980 Population"] > 80000000, "Country"]
top5
```

```python
popList.plot.scatter(x="Country", y = "1980 Population", alpha = 0.9)
plt.show()
```

```python
#box plot
popList.plot.box()
plt.show()
```

```python
axs = popList.plot.area(figsize = (1,20), subplots = True)
plt.show()
```
